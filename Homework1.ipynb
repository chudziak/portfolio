```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('avocado-updated-2020.csv')
df.tail(10)
```

```python
year = df.year.unique()
prices = []
for y in year:
    year_data = df[df.year==y]
    price = sum(year_data.average_price)/year_data.shape[0]
    prices.append(price)
price_data = pd.DataFrame({"Year":year,"Average_prices":prices})
price_data.head()
```

```python
fig = plt.figure(figsize=[16,8])
plt.bar(price_data.Year,price_data.Average_prices)
plt.xticks(rotation = 45)
plt.xlabel("Years")
plt.ylabel("Average Price")
plt.show()
```

```python
year = df.year.unique()
volumes = []
for y in year:
    year_data = df[df.year==y]
    volume = sum(year_data.total_volume)/year_data.shape[0]
    volume2=["%.2f" % i for i in volumes] ##https://www.pythonpool.com/python-float-to-string/
    ## ^^^^^ converts the float to a string in an array 
    volumes.append(volume)
volume_data = pd.DataFrame({"Year":year,"Total_Volume":volumes})
volume_data.head(6)
```

```python
fig = plt.figure(figsize=[16,8])
plt.bar(volume_data.Year,volume_data.Total_Volume)
plt.xticks(rotation = 45)
plt.ylabel("Total Volume (In Millions)")
plt.xlabel("Years")
plt.show()
```
